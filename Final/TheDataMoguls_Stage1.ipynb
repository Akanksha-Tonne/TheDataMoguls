```python
#reading the dataset
import pandas as pd
df1 = pd.read_csv('/kaggle/input/microsoft-malware-prediction/train.csv', dtype=dtypes)
```

```python
#describing the dataset
df1.describe()
```

```python
#PREPROCESSING

#1) DIMENTIONALITY REDUCTION

#based on missing value ratio
filtered_col = df1.columns[df1.isnull().mean()>0.5]
print(filtered_col)
print(len(df1.columns))
df=df1.drop(list(filtered_col),axis=1)
print(len(df.columns))
```

```python
#based on the variance of each coloumn
#An attribute with low variance not really separating data points in anyway.so we can remove that coloumn
l=[]
threshold=0.8
for i in df.columns:
    max_rfrequency = df[i].value_counts(normalize=True, dropna=False).values[0]#finding maximum value of the relative frequency of values in that column
    if(max_rfrequency>threshold):
        l.append(i)
print(l)
print(len(l))
df=df.drop(l,axis=1)
print(len(df.columns))
```

```python
print(df.dtypes) #printing datatype of each column
#figuring out which variables are categorical and which are numerical
#A variable having numeric type doesnot imply that it is quantitative,the numbers could represent categories/levels also
#from inspection of the dataset(nature and domain of the values an attribute can take ) and calculating unique values under each 
#column we can understand that  majority of the variables are  categorical  eventhough their datatype is numerical
#finding unique values of each column
for i in df.columns:
    print(i,len(df[i].unique()))
```

```python
#REPLACING MISSING VALUES IN EACH ROW 
#since majority of the columns are categorical we are replacing the missing values by mode
for column in df.columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
preprocessed=df
```

```python
#Since our dataset is huge and has a lot of categorical variable, Light GBM proves to be a very effective approach
#Light GBM(Gradient Boosting) is a tree based classifier that grows level wise rather than leaf wise
#it is very fast and works extremely well on large dataset without requiring to encode the categorical data
#majority of the columns in our dataset are categorical, so tested this model which gave us a estimate of the most important 
#which contributes maximum to the classification problem
import lightgbm as lgb,gc

# CREATE TRAIN AND VALIDATE
X_train = df.sample(frac=0.5)
Y_train = X_train['HasDetections']
X_val = df[ ~df.index.isin(X_train.index) ]
Y_val = X_val['HasDetections']
del X_train['HasDetections'], X_val['HasDetections'], df #removing 'has detections' from the data
x=gc.collect()

# TRAIN LGBM
model = lgb.LGBMClassifier(n_estimators=10000, colsample_bytree=0.7, objective='binary', num_leaves=32,
            max_depth=-1, learning_rate=0.1)
h=model.fit(X_train, Y_train, eval_metric='auc', eval_set=[(X_val, Y_val)], verbose=50,
            early_stopping_rounds=100)

# FEATURE IMPORTANCE
df1 = pd.DataFrame({"mean" : model.feature_importances_, "feature" : X_train.columns })
df1.sort_values("mean", inplace=True)
ax = df1.plot(x="feature", y="mean", kind='barh',color='green', figsize=(12, 20)
             , title='External Data LGBM Feature Importance')
```

```python
#since the version numbers in our dataset is associated with time we are using external datasets provided by the kaggle
temp=preprocessed
import numpy as np
from datetime import datetime, date, timedelta
date_AS = np.load('../input/malware-timestamps/AvSigVersionTimestamps.npy',allow_pickle=True)[()]

#maping the Microsoft's AvSigVersion (str) to a timestamp (datetime.datetime)
temp['date_AS'] = temp['AvSigVersion'].map(date_AS)

date_OS = np.load('../input/malware-timestamps-2/OSVersionTimestamps.npy',allow_pickle=True)[()]

#maping the Microsoft's Census_OSVersion (str) to a timestamp (datetime.datetime)
temp['date_OS'] = temp['Census_OSVersion'].map(date_OS)
print(len(temp))

#some computers have AvSigVersion = '0.0.0.0' and this key is not in the dictionary.so inorder to avoid NaN in the dataset we have to remove
#all the rows containing the value '0.0.0.0' for AvSugVersion
temp = temp[temp['AvSigVersion']!='0.0.0.0']
print(len(temp))
```

```python
# extracting only month information from date_AS column
AS_month=[]
for i in temp['date_AS']:
    k=str(i).split('-')
    if(len(k)>2):
        AS_month.append(int(k[1]))
    
#AS_month will contain list of all months for AvSigVersion
```

```python
%matplotlib inline      
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,8))
temp['AS_month']=AS_month
ax=sns.countplot(x="AS_month", hue="HasDetections", data=temp)
ax.set_title('Variation of detectioncount with month')
```

**The above plot shows malware detection count in each month. From the plot we can understand there is a sudden increase in the detection count during the months July, August and September compared to other months.**

```python
#extracting only year information from date_AS column
AS_year=[]
for i in temp['date_AS']:
    k=str(i).split('-')
    AS_year.append(int(k[0]))
    
#AS_month will contain list of all years for AvSigVersion
```
